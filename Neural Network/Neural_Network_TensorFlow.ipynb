```python
#Feature Scaling
def scaling(x):
    norm_1=tf.keras.layers.Normalization(axis=1)
    norm_1.adapt(x)
    x=norm_1(x)
    return x
```

```python
def repeat(x,reps_x=1000):
    x=np.tile(x,(reps_x,1))
    return x
```

```python
x=scaling(x)
#y=scaling(y)
# print(np.max(x[:,0]), np.min(x[:,0]))
# print(np.max(x[:,1]), np.min(x[:,1]))

x=repeat(x)
y=repeat(y)
#print(x.shape,y.shape)
```

```python
tf.random.set_seed(1234)
model=Sequential([
    tf.keras.Input(shape=(2,)),
    Dense(3,activation='sigmoid',name='layer_1'),
    Dense(1,activation='sigmoid',name='layer_2')
])
```

```python
model.summary()
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer='adam'
)
```

```python
model.fit(x,y,epochs=10)
```

```python
w1,b1=model.get_layer('layer_1').get_weights()
w1,b1
```

```python
x_test=np.array([
    [200,13.9],
    [200,17]
])
x_test=scaling(x_test)
```

```python
prediction=model.predict(x_test)
```
